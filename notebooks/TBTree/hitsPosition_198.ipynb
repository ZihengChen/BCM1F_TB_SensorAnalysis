```python
import pandas as pd
from pylab import *
%matplotlib inline
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_pickle("/Users/zihengchen/Documents/BRIL/data/TBTree/198.pkl").query("nTriples==1")
df = df.query("(VMECh2_pulsePosPassCuts & VMECh2_pulseAmpPassCuts) | (VMECh3_pulsePosPassCuts & VMECh3_pulseAmpPassCuts)").reset_index(drop=True)
```

```python
fig = plt.figure(figsize=(10,5),facecolor='w')
gs = gridspec.GridSpec(nrows=2, ncols=2, width_ratios=[1, 1])
fig.subplots_adjust(hspace=0)
ampBins = np.arange(0,41,1) 

ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(df.VMECh2_pulseAmp,ampBins,facecolor="grey",label="VMECh2")
ax1.grid(linestyle="--",alpha=0.4,color="gray")
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(df.VMECh3_pulseAmp, ampBins,facecolor="grey",label="VMECh3")
ax2.grid(linestyle="--",alpha=0.4,color="gray")
ax2.set_xlabel("pulse Amp [ADC]")

ax3 = fig.add_subplot(gs[:, 1])
xyBins = np.arange(-6,6,0.2)
x = np.resize(np.array([df.tposX_dut]),len(df))
y = np.resize(np.array([df.tposY_dut]),len(df))
ax3.hist2d(x,y,bins=(xyBins,xyBins),cmap="hot")
ax3.set_xlabel("x",fontsize=14)
ax3.set_ylabel("y",fontsize=14)
ax3.axis('equal')
ax3.set_xlim(-5,5)
ax3.set_ylim(-5,5)


from matplotlib.patches import Rectangle
currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((-4,-4), 8, 8, fill=None, color='w', alpha=1))
currentAxis.add_patch(Rectangle((-3,-.1), 5, 2.2, fill=None, color='w', alpha=1))

ax1.legend()
ax2.legend()
plt.savefig("198_empty.png",dpi=300)
```
