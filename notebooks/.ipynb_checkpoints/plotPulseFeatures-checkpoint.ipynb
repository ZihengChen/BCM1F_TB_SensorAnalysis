```python
import pandas as pd
from pylab import *
from TBTree_PulseAmpFitter import PulseAmpFitter

%matplotlib inline
dataDir = "../data/TBTree"
```

# 1. Fit One Run

```python
run = "198"
df = pd.read_pickle(dataDir + "/{}.pkl".format(run))
```

```python
backend, channel = "uTCA", 3
    
ch = df.query("{}Ch{}_pulsePosPassCuts & {}Ch{}_pulseAmpPassCuts".format(backend,channel,backend,channel)).reset_index(drop=True)
title = "Run {}, sensor-1 300V, {} CH{}, n={}".format(run, backend, channel, len(ch))
# fitting
v = "{}Ch{}_pulseAmp".format(backend,channel)
fitter = PulseAmpFitter(ch[v],binsize=1)
fitter.fitRange(30,90, None, True)
# plot
fitter.plot()
fitter.axes[0].set_title(title, fontsize=14)
```

# 2. Fit HVScan

```python
runMng = {
    "u"  :[50,100,150,200,250,300,350,400,500],               
    "run":["223","216","217","218","219","198","220","221","222"],
    "uTCACh2_fitRange":[(12,60),(20,80),(20,80),(20,80),(25,80),(25,80),(25,80),(25,80),(25,80)],
    "uTCACh3_fitRange":[(12,60),(20,80),(20,80),(20,80),(25,80),(25,80),(25,80),(25,80),(25,80)],
    "VMECh2_fitRange" :[( 6,40),(10,40),(10,40),(15,40),(15,40),(15,40),(15,40),(15,40),(15,40)],
    "VMECh3_fitRange" :[( 6,40),(10,40),(10,40),(15,40),(15,40),(15,40),(15,40),(15,40),(15,40)]}
```

```python
for run in runMng["run"]:
    df = pd.read_pickle(dataDir + "/{}.pkl".format(run))
    
    for backend in ["uTCA","VME"]:
        for channel in "23":
    
            ch = df.query("{}Ch{}_pulsePosPassCuts & {}Ch{}_pulseAmpPassCuts".format(backend,channel,backend,channel)).reset_index(drop=True)
            title = "Run {}, sensor-1 300V, {} CH{}, n={}".format(run, backend, channel, len(ch))
            # fitting
            v = "{}Ch{}_pulseAmp".format(backend,channel)
            fitter = PulseAmpFitter(ch[v],binsize=1)
            fitter.fitRange(30,90, None, True)
            # plot
            fitter.plot()
            fitter.axes[0].set_title(title, fontsize=14)
            plt.savefig("../plots/fitPulseAmp/run{}_{}Ch{}.png".format(run,backend,channel),dpi=300)
```
