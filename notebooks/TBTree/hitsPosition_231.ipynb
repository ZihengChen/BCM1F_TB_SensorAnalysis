```python
import pandas as pd
from pylab import *
%matplotlib inline
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_pickle("/Users/zihengchen/Documents/BRIL/data/TBTree/231.pkl").query("nTriples==1")
df = df.query("(VMECh0_pulsePosPassCuts & VMECh0_pulseAmpPassCuts) | (VMECh1_pulsePosPassCuts & VMECh1_pulseAmpPassCuts)").reset_index(drop=True)
```

```python
fig = plt.figure(figsize=(10,5),facecolor='w')
gs = gridspec.GridSpec(nrows=2, ncols=2, width_ratios=[1, 1])
fig.subplots_adjust(hspace=0)
ampBins = np.arange(0,41,1) 

ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(df.VMECh0_pulseAmp,ampBins,facecolor="grey",label="VMECh0")
ax1.grid(linestyle="--",alpha=0.4,color="gray")


ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(df.VMECh1_pulseAmp, ampBins,facecolor="grey",label="VMECh1")
ax2.grid(linestyle="--",alpha=0.4,color="gray")
ax2.set_xlabel("pulse Amp [ADC]")

ax3 = fig.add_subplot(gs[:, 1])
xyBins = np.arange(-6,6,0.2)
x = np.resize(np.array([df.tposX_dut]),len(df))
y = np.resize(np.array([df.tposY_dut]),len(df))
ax3.hist2d(x,y,bins=(xyBins,xyBins),cmap="hot")
ax3.set_xlabel("x",fontsize=14)
ax3.set_ylabel("y",fontsize=14)
ax3.axis('equal')
ax3.set_xlim(-5,5)
ax3.set_ylim(-5,5)

# slts = ["8<VMECh0_pulseAmp<15","VMECh0_pulseAmp>20","5<VMECh1_pulseAmp<10","VMECh1_pulseAmp>12"]
# colors = ["C1","C2","C0","C3"]

# subdfs = [df.query(slt) for slt in slts]
# amp1   = [subdf.VMECh0_pulseAmp for subdf in subdfs]
# amp2   = [subdf.VMECh1_pulseAmp for subdf in subdfs]
# ax1.hist(amp1,ampBins,color=colors, stacked=True)
# ax2.hist(amp2,ampBins,color=colors, stacked=True)
# alphas = [0.5,0.3,0.5,0.3]
# for i in range(len(slts)):
#     subdf = subdfs[i]
#     x = np.resize(np.array([subdf.tposX_dut]),len(subdf))
#     y = np.resize(np.array([subdf.tposY_dut]),len(subdf))
#     ax3.scatter(x,y,alpha=alphas[i],marker='.',s=1,c=colors[i])
    
# ax1.axvspan( 8, 15, lw=0, alpha=0.2, color=colors[0],label="VMECh0 low pulse")
# ax1.axvspan(20, 40, lw=0, alpha=0.2, color=colors[1],label="VMECh0 normal pulse")
# ax2.axvspan( 5, 10, lw=0, alpha=0.2, color=colors[2],label="VMECh1 low pulse")
# ax2.axvspan(12, 40, lw=0, alpha=0.2, color=colors[3],label="VMECh1 normal pulse")

from matplotlib.patches import Rectangle
currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((-4,-4), 8, 8, fill=None, color='w', alpha=1))
currentAxis.add_patch(Rectangle((-3,-0.75), 5, 2.75, fill=None, color='w', alpha=1))

ax1.legend()
ax2.legend()
plt.savefig("231_empty_hot.png",dpi=300)
```
