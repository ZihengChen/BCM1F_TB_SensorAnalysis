```python
import pandas as pd
from pylab import *
from tb_DataReader import EventFeaturesReader
from tb_PulseAmpFitter import PulseAmpFitter
import ROOT

%matplotlib inline
dataDir = "../data/eventFeatures"
```

```python
plt.figure(figsize=(6,5),facecolor='w')


fig, axes = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios':[1,1]},facecolor='w',figsize=(8,5))
fig.subplots_adjust(hspace=0.2)

ax = axes[0]
v,a,b,step = "pulseAmp_knElectron",0.1,60,1

run,backend,channel = "198","VME",3
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),facecolor="#FFBD22",label="{}: Normal Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)
run,backend,channel = "243","VME",3
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),histtype="step",lw=1.5, color="C2",label="{}: LED Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)
run,backend,channel = "270","VME",1
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),histtype="step",lw=1.5, color="C0",label="{}: Inclined Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)

ax.legend()
ax.set_title("Sensor-1 300V, Selection and Calibration applied")
ax.grid(linestyle="--",alpha=0.4,color="gray")

ax = axes[1]
run,backend,channel = "198","ubcm",3
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),facecolor="#FFBD22",label="{}: Normal Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)
run,backend,channel = "243","ubcm",3
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),histtype="step",lw=1.5, color="C2",label="{}: LED Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)
run,backend,channel = "270","ubcm",1
ch = EventFeaturesReader(dataDir,"000"+run,backend).getChannel(channel)
ax.hist(ch[v],bins=np.arange(a,b,step),histtype="step",lw=1.5, color="C0",label="{}: Inclined Run #{} ch{}, n={}".format(backend,run,channel,len(ch)),normed=True)


ax.legend()
ax.grid(linestyle="--",alpha=0.4,color="gray")
ax.set_xlabel(r"Pulse Amplitude [$10^3 e$]")


plt.savefig("../plots/runStudy/senser1.png",dpi=300)
```
