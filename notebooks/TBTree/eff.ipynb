```python
import pandas as pd
from pylab import *
%matplotlib inline
import matplotlib.gridspec as gridspec
run = 198
df = pd.read_pickle("/Users/zihengchen/Documents/BRIL/data/TBTree/{}.pkl".format(run)).query("nTriples==1")
df.reset_index(drop=True,inplace=True)
```

```python
def getTPosHist(dataframe,xyBins):
    x = np.resize(np.array([dataframe.tposX_dut]),len(dataframe))
    y = np.resize(np.array([dataframe.tposY_dut]),len(dataframe))
    h,bx,by = np.histogram2d(x,y,bins=(xyBins,xyBins))
    return h

xyBins = np.arange(-5,5.1,0.1)
cutCol = [c for c in df.columns if "pulseAmpPassCuts" in c]
slt    = df[cutCol].sum(axis=1)>0
h      = getTPosHist(df,xyBins)
hslt   = getTPosHist(df[slt],xyBins)
```

```python
eff = hslt/(h+1e-5)
```

```python
fig = plt.figure(figsize=(5,5),facecolor='w')
# plt.pcolor(xyBins,xyBins,eff.T)
plt.imshow(eff.T, cmap='viridis',vmin=0, vmax=1,
              extent=[xyBins.min(), xyBins.max(), xyBins.min(), xyBins.max()],
              interpolation='nearest', origin='lower',aspect="equal")
# plt.axis('equal')
from matplotlib.patches import Rectangle
currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((-4,-4), 8, 8, fill=None, color='w', alpha=1))
currentAxis.add_patch(Rectangle((-3,-.1), 5, 2.2, fill=None, color='w', alpha=1))
plt.xlim(-4,3)
plt.ylim(-2,3)
plt.xlabel("x [mm]")
plt.ylabel("y [mm]")
plt.colorbar(shrink=0.6)
plt.title("Run {}: ".format(run) + r"$\epsilon$ = P ( $A>5.1\sigma_{bs}$ | $n_{tri}=1$ )")
plt.savefig("{}_eff.png".format(run),dpi=300)
```
