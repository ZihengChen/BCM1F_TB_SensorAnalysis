```python
import pandas as pd
from pylab import *
%matplotlib inline
import matplotlib.gridspec as gridspec
```

```python
dfall = pd.read_pickle("/Users/zihengchen/Documents/BRIL/data/TBTree/198.pkl")#.query("nTriples==1")

fig = plt.figure(figsize=(10,5),facecolor='w')
from matplotlib.patches import Rectangle


plt.subplot(121)
df = dfall.query("(VMECh2_pulsePosPassCuts & VMECh2_pulseAmpPassCuts) | (VMECh3_pulsePosPassCuts & VMECh3_pulseAmpPassCuts)")
amp1 = df.VMECh2_pulseAmp/df.VMECh2_baselineStd
amp2 = df.VMECh3_pulseAmp/df.VMECh3_baselineStd
plt.scatter(amp1[amp2<5.1],amp2[amp2<5.1],color="C0",s=1,alpha=0.2)
plt.scatter(amp1[amp1<5.1],amp2[amp1<5.1],color="C0",s=1,alpha=0.2)
slt = np.logical_and(amp1>5.1 ,amp2>5.1)
plt.scatter(amp1[slt],amp2[slt],color="C1",marker="s",s=10,alpha=1)

plt.axvline(5.1,color="k",alpha=1,lw=2,linestyle="--")
plt.axhline(5.1,color="k",alpha=1,lw=2,linestyle="--")
plt.axvspan(0,5.1,color="C0",alpha=0.2)
plt.axhspan(0,5.1,color="C0",alpha=0.2)
plt.legend()
plt.xlim(0,40)
plt.ylim(0,40)
plt.xlabel(r"VME_Ch2_pulseAmp [$\sigma_{bs}$]")
plt.ylabel(r"VME_Ch3_pulseAmp [$\sigma_{bs}$]")
plt.title("Run 198 sensor-1 VME")



plt.subplot(122)
df = dfall.query("(uTCACh2_pulsePosPassCuts & uTCACh2_pulseAmpPassCuts) | (uTCACh3_pulsePosPassCuts & uTCACh3_pulseAmpPassCuts)")
amp1 = df.uTCACh2_pulseAmp/df.uTCACh2_baselineStd
amp2 = df.uTCACh3_pulseAmp/df.uTCACh3_baselineStd
plt.scatter(amp1[amp2<5.1],amp2[amp2<5.1],color="C0",s=1,alpha=0.2)
plt.scatter(amp1[amp1<5.1],amp2[amp1<5.1],color="C0",s=1,alpha=0.2)
slt = np.logical_and(amp1>5.1 ,amp2>5.1)
plt.scatter(amp1[slt],amp2[slt],color="C1",marker="s",s=10,alpha=1)

plt.axvline(5.1,color="k",alpha=1,lw=2,linestyle="--")
plt.axhline(5.1,color="k",alpha=1,lw=2,linestyle="--")
plt.axvspan(0,5.1,color="C0",alpha=0.2)
plt.axhspan(0,5.1,color="C0",alpha=0.2)

plt.legend()
plt.xlim(0,40)
plt.ylim(0,40)
plt.xlabel(r"uTCA_Ch2_pulseAmp [$\sigma_{bs}$]")
plt.ylabel(r"uTCA_Ch3_pulseAmp [$\sigma_{bs}$]")
plt.title("Run 198 sensor-1 uTCA")

plt.savefig("198_crosstalk.png",dpi=300)
```
